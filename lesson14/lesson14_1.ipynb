```python
odd = [1,3,5,7,9]
even = [2,4,6,8,10]

#list使用 + 運算子
odd + even
#list的加法 是將兩個list合成一個新的 list

#如果想要讓 list 內的元素數值相加 則必須將個數值一個一個取出處理
```

```python
#使用numpy
import numpy as np

odd_array = np.array(odd)
even_array = np.array(even)

#numpy.ndarray 使用數學運算子
#numpy.ndarray 可以做陣列的運算
#可以省略 list 使用 for ... in ... 的取值計算方式
print (odd_array + even_array)
print (odd_array * even_array)
print (odd_array / even_array)
print (odd_array - even_array)
```

## numpy的數學運算

```python
t = np.arange(0 , 2.5 , 0.1)
print(t)
print(t.size)
print(t.ndim)
print(t.dtype)
print(t.shape)
```

```python
import math
import matplotlib.pyplot as plt

t = np.arange(0 , 2.5 , 0.1)
y1 = np.sin(t * math.pi)
y2 = np.sin(t * math.pi + math.pi * 1/2)
y3 = np.sin(t * math.pi - math.pi * 1/2)

plt.plot(t,y1,marker='*',color='b',linestyle='--')
plt.plot(t,y2,marker='v',color='r',linestyle='-.')
plt.plot(t,y3,marker='v',color='g',linestyle=':')
plt.savefig('a.png')
plt.show()
plt.close()
```
