```python
import numpy as np
import pandas as pd
data = {
    'apples':[2,3,0,1],
    'oranges':[0,3,7,2]
}

pd.DataFrame(data,index=('一','二','三','四'))
```

```python
data = np.random.randint(50,101,size=(5,5))
df = pd.DataFrame(data,
                  index=['學生一','學生二','學生三','學生四','學生五'],
                  columns=['科目一','科目二','科目三','科目四','科目五'])
df
```

```python
#依據欄位名稱的選取
df['科目一']    #series
df[['科目一','科目二']]     #dataframe
```

```python
#使用numpy的統計方法來做計算
sum = np.sum(df,axis=1)
mean = np.mean(df,axis=1)
df['總分'] = sum
df['平均'] = mean
#加入了總分與平均 所以再做計算的話會把總分與平均加入計算
```

```python
df
```

```python
#依據列的選取
#.loc[] >> 名稱
#.iloc[] >> 按照索引 index

df.loc['學生一']
df.loc[['學生一','學生二']]
df.iloc[0]
df.iloc[:2]
df.iloc[0,:5]
df.iloc[:2,:5]
```
