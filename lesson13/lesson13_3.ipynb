```python
from matplotlib.font_manager import fontManager
import matplotlib as mpl

fontManager.addfont('ChineseFont.ttf')
mpl.rc('font',family='ChineseFont')
```

```python
import matplotlib.pyplot as plt

student1 = {'':78, '':80, '':92, '':75, '':85}
student2 = {'':90, '':75, '':98, '':65, '':75}

students = list(student1.keys())
student1_value = list(student1.values())
student2_value = list(student2.values())

figure = plt.figure(figsize=[8,5],dpi=200,facecolor='#cccccc')
axes = figure.add_subplot(1,1,1)
axes.plot(students,student1_value,marker='o',label='')
axes.plot(students,student2_value,marker='o',linestyle='-.',label='')
axes.set_title('  ')
axes.set_xlabel('')
axes.set_ylabel('')
# Y  alpha 0 ~ 1
axes.grid(axis='y',color='pink',alpha=1,linestyle=':')
axes.legend()
xlocations = axes.get_xticks()
for index in range(len(xlocations)) :
    x = xlocations[index]
    student1y = student1_value[index]
    student2y = student2_value[index]
    # print('x:',x)
    # print('s1_y',student1y)
    # print('s2_y',student2y)
    # print('='*10)
    #  =
    axes.annotate(str(student1y),(x-0.05,student1y+1))
    if student2y == 98:
        axes.annotate(str(student2y),(x,student2y),xytext=(x+0.3, student2y+3), arrowprops=dict(facecolor='red', shrink=0.05))
    else :
        axes.annotate(str(student2y),(x-0.05,student2y+1))
# x y  
axes.set_xlim(-0.5,4.5)
axes.set_ylim(60,105)
plt.show()
```
