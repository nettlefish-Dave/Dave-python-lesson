```python
#透過wget下載字形
import wget
wget.download("https://github.com/roberthsu2003/machine_learning/raw/refs/heads/main/source_data/ChineseFont.ttf")
```

```python
import matplotlib.pyplot as plt
#如果想要使用中文標題 需載入字形管理工具
from matplotlib.font_manager import fontManager
import matplotlib as mpl
#使用fontmanager加入中文字形
fontManager.addfont('ChineseFont.ttf')
#將字形加入matplotlib的字形庫
mpl.rc('font',family='ChineseFont')
#快速建立畫布 figure 與子圖 subplot
figure, axes = plt.subplots()
x = (0,1,4,5,7,10)
y = (3,4,5,6,9,13)
axes.plot(x,y,marker='v',label='學生一',linestyle='--',color='r',markersize=10,linewidth=2)
axes.set_title('折線圖')
axes.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
#如果想要使用中文標題 需載入字形管理工具
from matplotlib.font_manager import fontManager
import matplotlib as mpl
#使用fontmanager加入中文字形
fontManager.addfont('ChineseFont.ttf')
#將字形加入matplotlib的字形庫
mpl.rc('font',family='ChineseFont')
#快速建立畫布 figure 與子圖 subplot
figure, axes = plt.subplots()
x = (0,1,2,3,4)
y = (3,4,5,6,7)
axes.plot(x,y,marker='v',label='學生一',linestyle='--',color='r',markersize=10,linewidth=2)
axes.set_title('折線圖')
axes.set_xticks([0,1,2,3,4],['零','一','二','三','四'])
axes.legend()
plt.show()
```
