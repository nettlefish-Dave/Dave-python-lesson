### 擴充套件 site package 放外部套件
- requirements.txt 裡面紀錄使用到的套件
- 使用 pip install -r requirements.txt 來安裝所有文件中的套件
- 使用 pip list 可以查詢所有已經安裝的擴充套件
- 使用 pip show 套件名稱 可以詳細查詢該套件資訊

```python
import requests
from requests import Response

url:str = 'https://raw.githubusercontent.com/nettlefish-Dave/Dave-python-lesson/refs/heads/main/lesson10/students.csv'
response:Response = requests.get(url)
if response.status_code == 200:
    content:str = response.text
    print('下載成功')
```

```python
import io
import csv
# 將 str 物件轉為 file 物件
file = io.StringIO(content)
reader = csv.DictReader(file)
data = [row for row in reader]
file.close()
data
```
